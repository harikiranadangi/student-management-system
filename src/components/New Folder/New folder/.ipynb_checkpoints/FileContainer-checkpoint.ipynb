```python
import os

def list_files_and_folders(path, exclude):
    files = []
    folders = []
    
    for root, dirs, file_names in os.walk(path):
        # Exclude specified folders
        dirs[:] = [d for d in dirs if d not in exclude]
        
        # Append remaining folder names
        for dir_name in dirs:
            folders.append(os.path.join(root, dir_name))
        
        # Append file names
        for file_name in file_names:
            files.append(os.path.join(root, file_name))
    
    return files, folders

# Replace this with the path to your folder
folder_path = r"D:\GITHUB\student-management-system"

# List of folders to exclude
excluded_folders = [".next", ".vscode", "MindMap", "NewFolder", "node_modules", ".git"]

# Get files and folders
files, folders = list_files_and_folders(folder_path, excluded_folders)

# Print folder and file names
print("Folders:")
for folder in folders:
    print(folder)

print("\nFiles:")
for file in files:
    print(file)

# Print counts
print(f"\nTotal Folders: {len(folders)}")
print(f"Total Files: {len(files)}")
```
