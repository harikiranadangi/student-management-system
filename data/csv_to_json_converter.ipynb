```python
import pandas as pd
import numpy as np

# Load the CSV file
df = pd.read_csv(r'D:\GITHUB\student-management-system\data\students_data.csv')
print('Total Students:',df.shape[0])
df.head()
```

```python
df.columns = df.columns.str.lower()
```

```python
df[df['dob'].isnull()]
```

```python
# Convert 'id' and 'phone' to string to match Prisma schema
df['id'] = df['admissionno'].astype(str)

df['username'] ='S' + df['id']

df['name'] = df['student_name'].astype(str)

df['surname'] = df['student_name'].str.split().str[-1]

df["parentName"] = df["fathername"].fillna(df["mothername"]).fillna('NA')

df['email'] = df['username'] + '@kotakschool.com'

df['phone'] = df['mobile'].astype(str)

# Convert 'address' to string and handle missing values
df["address"] = ("Visakhapatnam")

df['img'] = np.nan

df['bloodType'] = 'Under Investigation'

df['classId'] = df['classno']

df['createdAt'] = pd.to_datetime("today").date()

df['deletedAt'] = np.nan

df_cleaned = df[['id', 'username', 'name', 'surname', 'parentName', 'email', 'phone',
       'address', 'img', 'bloodType', 'gender', 'dob', 'createdAt',
       'deletedAt', 'classId']].copy()

df.tail(1)
```

```python
print(df.shape)
df['dob'] = df['dob'].astype(str)
print(df[['id', 'username', 'dob']].head(10))

```

```python
df_cleaned.to_csv(r'D:\GITHUB\student-management-system\data\student_data.csv')

# Load the CSV file
print('Total Students:',df_cleaned.shape[0])
df_cleaned.head()
```
