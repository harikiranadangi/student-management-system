```python
import pandas as pd

# Load the CSV file
df = pd.read_csv(r'D:\GITHUB\student-management-system\NewFolder\students_template.csv')

df


```

```python
# Convert 'id' and 'phone' to string to match Prisma schema
df['id'] = df['id'].astype(str)
df['phone'] = df['phone'].astype(str)

# Convert 'address' to string and handle missing values
df["address"] = df["address"].astype(str).fillna("unknown")

# Ensure dob is never missing and convert to ISO-8601 format
df["dob"] = df["dob"].replace(["00-00-0000", "00-00-00", None, ""], "01-01-2000")
df["dob"] = df["dob"].fillna("01-01-2000")  # Fallback for NaN values
df['dob'] = pd.to_datetime(df['dob'], format='%d-%m-%Y', errors='coerce').dt.strftime('%Y-%m-%dT00:00:00.000Z')

# Fill missing parentName with the student's name
df['parentName'] = df['parentName'].fillna(df['name'])
```

```python

df_json = df.copy()

# Save as JSON in the correct format
df_json.to_json(r'D:\GITHUB\student-management-system\prisma\students.json', orient='records')

print("JSON file saved successfully with valid dob format!")
```
